```python
import pandas as pd
import numpy as np
```

```python
# If not already installed, do: pip install pandas fastparquet
URL_DATA = 'https://storage.data.gov.my/pricecatcher/lookup_item.parquet'

df = pd.read_parquet(URL_DATA)
if 'date' in df.columns: df['date'] = pd.to_datetime(df['date'])

lookup = df

lookup
```

```python
# If not already installed, do: pip install pandas fastparquet
URL_DATA = 'https://storage.data.gov.my/pricecatcher/lookup_premise.parquet'

df = pd.read_parquet(URL_DATA)
if 'date' in df.columns: df['date'] = pd.to_datetime(df['date'])

premise = df

premise
```

```python
# If not already installed, do: pip install pandas fastparquet

URL_DATA = 'https://storage.data.gov.my/pricecatcher/pricecatcher_2025-01.parquet'

df = pd.read_parquet(URL_DATA)
if 'date' in df.columns: df['date'] = pd.to_datetime(df['date'])

price_catcher = df
price_catcher
```

```python
merge = lookup.merge(price_catcher,on='item_code',how='left')
full_merge = merge.merge(premise,on='premise_code',how='left')
full_merge
```

```python
full_merge.isna().sum()
full_merge = full_merge.dropna()
full_merge
```

```python
ayam_list = list(lookup[lookup['item_category']=='AYAM'].iloc[:,0])
```

```python
ayam_cat = full_merge[full_merge['item_code'].isin(ayam_list)]
ayam_cat
```

```python
ayam_cat.groupby(['item_code','item','item_category']).agg(avg=('price','mean'),
                                                          max=('price','max'),
                                                          min=('price','min'))
```

```python
var_012525 = full_merge[full_merge['date']=='2025-01-25']
var_tgg = var_012525[var_012525['district']=='Kuala Terengganu']
var_ayamb = var_tgg[var_tgg['item_code']==1]
var_ayamb.sort_values(by=['price'])
```

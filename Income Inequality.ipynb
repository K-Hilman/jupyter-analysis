```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
URL_DATA = 'https://storage.dosm.gov.my/hies/hh_inequality_state.parquet'

df = pd.read_parquet(URL_DATA)
if 'date' in df.columns: df['date'] = pd.to_datetime(df['date'])

print(df)
```

```python
#Change to dt.
df['year'] = df['date'].dt.year
df
```

```python
df['state'].value_counts()
```

```python
year_1 = 1984
year_2 = 2022

df1 = df[df['year'] == year_1].sort_values('gini',ascending=False)
df2 = df[df['year'] == year_2].sort_values('gini',ascending=False)
df1
```

```python
fig, axes = plt.subplots(1,2,figsize=(22,8))

fig.suptitle(f'Gini coefficient by states {year_1} vs {year_2}')

sns.barplot(ax=axes[0], data = df1, x='gini',y='state', color='cyan')
sns.barplot(ax=axes[1], data = df2, x='gini',y='state', color='cyan')
```

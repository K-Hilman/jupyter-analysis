```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# If not already installed, do: pip install pandas fastparquet
import pandas as pd

URL_DATA = 'https://storage.dosm.gov.my/demography/marriages_age.parquet'

df = pd.read_parquet(URL_DATA)
if 'date' in df.columns: df['date'] = pd.to_datetime(df['date'])

marriage = df
print(marriage)
```

```python
marriage['year'] = marriage['date'].dt.year
marriage = marriage.drop(columns='date')
marriage
```

```python
marriage_piv_male = marriage[marriage['sex']=='male'].pivot(index='year',values='abs',columns='age').drop(columns='all').reset_index()
marriage_piv_male.to_csv('marriage_piv_male.csv')
marriage_piv_male['gender'] = 'male'
marriage_piv_male
```

```python
marriage_piv_female = marriage[marriage['sex']=='female'].pivot(index='year',values='abs',columns='age').drop(columns='all').reset_index()
marriage_piv_female.to_csv('marriage_piv_female.csv')
marriage_piv_female['gender'] = 'female'
marriage_piv_female
```

```python
marriage_piv = pd.concat([marriage_piv_male,marriage_piv_female])
marriage_piv.to_csv('marriage_piv.csv')
marriage_piv
```
